```python
from datetime import datetime, date, timedelta
try:
    from dateutil.easter import easter
except:
    !pip -q  install python-dateutil
    from datetutil.easter import easter
```

```python
non_settlement_2026_2028_us=holidays.US(years=[2026,2028])
non_settlement_2026_2028_us[easter(2026)-timedelta(days=2)]='Good Friday'
non_settlement_2026_2028_us[easter(2027)-timedelta(days=2)]='Good Friday'
```

```python
july_4_2026=date(2026,7,4)
july_4_2028=date(2028,7,4)
for july_4 in (july_4_2026,july_4_2028):
  adjusted_date=july_4
  while adjusted_date.weekday()>4 or adjusted_date in non_settlement_2026_2028_us:
    adjusted_date+=timedelta(days=1)
    if adjusted_date.year not in non_settlement_2026_2028_us:
      non_settlement_2026_2028_us.update(years=date.years)
  display(adjusted_date)
```
